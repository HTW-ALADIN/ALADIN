```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Beispiel Daten - hier können Sie Ihre eigenen Daten eingeben
np.random.seed(0)
num_points = 10  # Anzahl der Datenpunkte
num_dimensions = 2  # Anzahl der Dimensionen der Datenpunkte
X = np.random.random_sample([num_points, num_dimensions])

# Initialisierung: Jeder Datenpunkt ist ein Cluster
clusters = [[i] for i in range(X.shape[0])]

# Schleife für jeden Schritt der hierarchischen Clusteranalyse
for step in range(X.shape[0] - 1):
    print(f"\n--- Schritt {step + 1} ---")
    print("Aktuelle Cluster:", clusters)

    # Berechnung der Abstandsmatrix mit Ward-Methode
    linkage_matrix = linkage(X, 'ward')

    # Ausgabe der Abstandsmatrix
    print("Abstandsmatrix nach Linkage:")
    print(linkage_matrix)

    # Tabelle mit den Abständen zwischen den Clustern erstellen
    distances = []
    for i in range(len(linkage_matrix)):
        cluster1, cluster2, distance, _ = linkage_matrix[i]
        cluster1, cluster2 = int(cluster1), int(cluster2)  # Konvertiere zu ganzzahligen Indizes
        if cluster1 < len(clusters) and cluster2 < len(clusters):
            distances.append([cluster1, cluster2, distance, len(clusters[cluster1]), len(clusters[cluster2])])

    # Tabelle als DataFrame erstellen und ausgeben
    #columns = ['Cluster 1', 'Cluster 2', 'Abstand', 'Größe Cluster 1', 'Größe Cluster 2']
    #distance_table = pd.DataFrame(distances, columns=columns)
    #print("Tabelle mit Abständen zwischen Clustern:")
    #print(distance_table)

    # Streudiagramm zeichnen
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    plt.figure(figsize=(8, 6))
    for idx, cluster in enumerate(clusters):
        for point_idx in cluster:
            plt.scatter(X[point_idx, 0], X[point_idx, 1], color=colors[idx % len(colors)], label=f'Cluster {idx}')
    plt.title('Hierarchische Clusteranalyse - Schritt {}'.format(step + 1))
    plt.legend()
    plt.show()

    # Wenn nur noch ein Cluster übrig ist, breche die Schleife ab
    if len(clusters) == 1:
        break

    # Identifiziere die beiden nächsten Cluster
    cluster1, cluster2 = int(linkage_matrix[step][0]), int(linkage_matrix[step][1])

    # Überprüfe, ob die Cluster existieren, bevor sie zusammengeführt werden
    if cluster1 < len(clusters) and cluster2 < len(clusters):
        new_cluster = clusters[cluster1] + clusters[cluster2]
        del clusters[max(cluster1, cluster2)]
        del clusters[min(cluster1, cluster2)]
        clusters.append(new_cluster)

# Ausgabe des finalen Clusters
print(f"\n--- Schritt {X.shape[0]} ---")
print("Finale Cluster:", clusters)

# Dendrogramm erstellen und anzeigen
plt.figure(figsize=(10, 7))
plt.title('Hierarchische Clusteranalyse - Methode: Ward')
dendrogram(linkage_matrix)
plt.show()
```
